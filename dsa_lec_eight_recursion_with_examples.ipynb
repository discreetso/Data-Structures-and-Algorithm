```python
# factorial
def fact(n):
    # base case
    if n <= 1:
        return 1

    # induction case
    return n * fact(n - 1)
```

```python
for i in range(1, 7):
    print("Factorial of ", i, " is: ", fact(i))
```

```python
# fib (recursively)
def fib(n):

    # base case
    if n <= 1:
        return 1

    # induction case
    return fib(n - 1) + fib(n - 2)           # more stuff after recursive call
```

```python
%timeit fib(20)
```

```python
# fib (iteratively)
def fib(n):
    a, b = 0, 1                   # disagreement between the value of a: 0 or 1

    for i in range(n):
        a, b = b, a + b
    return a
```

```python
fib(5)
```

```python
# tail recursion
def fib(n, a = 0, b = 1):
    if n == 0:
        return a 

    return fib(n - 1, b, a + b)           # since everything is done at the last call, this is called 'tail recursion'
```

```python
fib(8)
```

```python
def tower_of_hanoi(levels = 3):
    move_tower(levels, 'A', 'C', 'B')        # move n-level tower from A to C using B as aux

def move_tower(l, fr, to, ax):
    if l == 1:                             # base case
        print_move(l, fr, to)
        return

    # induction case
    move_tower(l - 1, fr, ax, to)
    print_move(l, fr, to)
    move_tower(l - 1, ax, to , fr)

def print_move(l, fr, to):
    print("Move ", l, " from ", fr, " to ", to)
```

```python
tower_of_hanoi(2)
```

```python
# sum over a list
```

```python
def sum_list(l):
    sum = 0

    for i in l:
        sum += i
    return sum
```

```python
l = [1, 2, 3, 4, 5]
```

```python
sum_list(l)
```

```python
%timeit sum_list([1, 2, 3, 4, 5])
```

```python
def sum_list_recursive(l):
    if len(l) == 0: return 0            # base case: sum of empty list is 0

    return l[0] + sum_list_recursive(l[1:])        # induction case: head + sum of the tail
```

```python
sum_list_recursive(l)
```

```python
%timeit sum_list_recursive()
```
