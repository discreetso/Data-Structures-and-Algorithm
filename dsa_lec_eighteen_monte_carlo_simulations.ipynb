```python
import random as r
from math import sqrt
```

```python
def distance(a, b):
    return sqrt(a**2 + b**2)

def estimate_pi(darts):
    inside = 0

    for i in range(darts):
        x = r.random()
        y = r.random()

        if distance(x, y) < 1.0:
            inside += 1

    pi = inside/darts * 4
    return pi
```

```python
num_sim = int(1e+8)
pi = estimate_pi(num_sim)
print(pi)
```

```python
# random walk (drunken walk)
import random

def get_action():
    return random.choice( [(0, 1), (0, -1), (1, 0), (-1, 0)] )

def random_walk(n):
    """ Return path after 'n' block random walk """
    path = []
    x, y = 0, 0

    for i in range(n):
        (dx, dy) = get_action()          # get step deltas

        x = x + dx
        y = y + dy

        path.append( (x, y) )          # create new tuple and save in the path
    return path
```

```python
walk = random_walk(200)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(num = None, figsize = (15, 10))
point_color = list(range(200))
x_numbers = [x[0] for x in walk]
y_numbers = [y[1] for y in walk]
plt.scatter(x_numbers, y_numbers, marker='.', s=50, c=point_color)
plt.show()
```
