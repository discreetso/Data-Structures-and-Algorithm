```python
# selection sort: start from the first index, take i (0) as the min_idx, 
# then find the minimum by comparing all the other items starting from i+1 to n (length of elements), 
# when it found, swap it with the first index element means move that minimun element to the first position
# then find the second minimum and the third minimum and so on.
```

```python
def selection_sort(l):
    n = len(l)

    # for each element in the list (starting from left)
    for i in range(n):
        min_idx = i                # find the minimum

        # ---- in the *rest* of the list
        for j in range(i+1, n):
            if l[j] < l[min_idx]:
                min_idx = j

        # swap the minimum with current element, now we have (sorted stuff till i)
        l[i], l[min_idx] = l[min_idx], l[i]
```

```python
l = [1, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
print("unsorted list: ", l)
selection_sort(l)
print("sorted with the selection sort: ", l)
```

```python
# quick sort: it is a very fast sorting algorithm as suggested by its name. But it is also counter-intuitive.
# it is a recurrsion based algorithm
```

```python
import random

def qsort(l, fst, lst):
    if fst >= lst: return 
        
    i, j = fst, lst
    pivot = l[random.randint(fst, lst)]
    
    while i <= j:
        while l[i] < pivot: i += 1
        while l[j] > pivot: j -= 1
        
        if i <= j:
            l[i], l[j] = l[j], l[i]
            i, j = i+1, j-1

    qsort(l, fst, j)
    qsort(l, i, lst)    
```

```python
l = [1, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
print("unsorted list: ", l)
qsort(l, 0, len(l)-1)
print("sorted with the quick sort: ", l)
```
