```python
# bubble sorting 

def bubble_sort(lst):
    n = len(lst)
    # print(n)

    # outer loop goes over the whole thing n times
    # because each time one 'highest' will move to the end
    swapped = False
    for i in range(n):
        # try to bubble the higher up one
        for j in range(0, (n-i) - 1):
            # compare pairs, move higher one up (the highest one will move to the end this pass, reducing the range for the next pass)
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
                swapped = True
        if not True:
            break
```
