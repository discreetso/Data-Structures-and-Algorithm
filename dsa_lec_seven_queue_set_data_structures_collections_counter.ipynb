```python
class Queue:
    def __init__(self):
        self.size = 5 
        self.q = list(range(self.size))              #some dummy values
        
        self.i = 0
        self.o = 0
        
        self.is_empty = True
        self.is_full = False
```

```python
q = Queue()
```

```python
q
```

```python
def _inc(self, idx):
    if idx + 1 == self.size:
        return 0
    else:
        return idx + 1

Queue._inc = _inc
```

```python
def enqueue(self, val):
    if self.is_full:
        raise IndexError("Cannot enqueue, list is full")

    self.q[self.i] = val
    self.i = self._inc(self.i)

    if self.i == self.o:
        self.is_full = True
    self.is_empty = False

Queue.enqueue = enqueue
```

```python
def dequeue(self):
    if self.is_empty:
        raise IndexError("Cannor dequeue, list is empty")

    ret = self.q[self.o]
    self.o = self._inc(self.o)

    if self.o == self.i:
        self.is_empty = True
    self.is_full = False
    return ret

Queue.dequeue = dequeue
```

```python
def __str__(self):
    return str(self.q) + ' , in: ' + str(self.i) + ' , out: ' + str(self.o)

Queue.__str__ = __str__
```

```python
q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)
q.enqueue(50)

# q.enqueue(35)

print(q)
```

```python
# All and Any
# you can use all and any if you have a lot of conditions -- all or some of which need to be satisfied
```

```python
a = 5
b = 4

if a > 2 and a < 10 and b > 0 and b < 10:
    print("All constraints satisfied")
```

```python
a = 5
b = 4
constraints = { a > 2, a < 10, b > 0, b < 10 }
```

```python
if all(constraints):
    print("All constraints satisfied")
```

```python
if any(constraints):
    print("At least one constraint satisfied")
```

```python
# Quantifiers (for all) and (there exists (for some))
```

```python
X = {10, 20, 30, 40, 50}
all([x < 100 for x in X])      # same as 'for all'
```

```python
any([x < 50 for x in X])      # same as 'there exists'
```

```python
# You can also have an arbitrary function computed in the scope
```

```python
def is_div_by_20(n):
    return n % 20 == 0    
```

```python
all([is_div_by_20(x) for x in X]) 
```

```python
any([is_div_by_20(x) for x in X]) 
```

```python
from collections import Counter
cnt = Counter()
for word in ['red', 'green', 'red', 'blue', 'green', 'red']:
    cnt[word] += 1
```
