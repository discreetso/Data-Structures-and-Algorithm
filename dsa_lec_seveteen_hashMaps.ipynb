```python
class HashMap:
    def __init__(self):
        self.size = 10
        self.map = [None] * self.size
```

```python
def _get_hash(self, key):
    return key % self.size

HashMap._get_hash = _get_hash
```

```python
def add(self, key, value):
    key_hash = self._get_hash(key)
    key_value = [key, value]
    
    # insert or update: 'upsert'
    self.map[key_hash] = [key_value]        # notice the double list, we will get to it in the next approach
    return True

HashMap.add = add
```

```python
def get(self, key):
    key_hash = self._get_hash(key)
    if self.map[key_hash] is not None:
        for pair in self.map[key_hash]:
            if pair[0] == key:
                return pair[1]

    raise KeyError(str(key))

HashMap.get = get
```

```python
def __str__(self):
    ret = ' '

    for i, item in enumerate(self.map):
        if item is not None:
            ret += str(i) + ': ' + str(item) + '\n'

    return ret

HashMap.__str__ = __str__
```

```python
h = HashMap()

h.add(17, 'seveteen')
h.add(26, 'twenty six')
h.add(35, 'thirty five')
h.add(26, 'twenty six updated')
```
