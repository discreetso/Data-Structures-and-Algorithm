```python
# Heap Data Structure - MaxHeap and MinHeap
```

```python
def heapify(lst, n, root):
    """ heapify the root element of the lst which has n elements in total """
    l = 2*root + 1         # gives the left node index
    r = 2*root + 2         # gives the right node index

    largest = root

    if l < n and lst[l] > lst[largest]:
        largest = l
    if r < n and lst[r] > lst[largest]:
        largest = r

    if largest != root:
        lst[root], lst[largest] = lst[largest], lst[root]
        heapify(lst, n, largest)
```

```python
def build_heap(lst):
    """ construct heap from any list by repeatedly applying heapify from bottom non-leaf nodes. """
    n = len(lst)
    for i in reversed( range(n // 2) ):            # rest must be leaves, no need to heapify
        # print("Heapify: ", lst[i])
        heapify(lst, n, i)
```

```python
heap = [100, 5, 3, 2, 8, 15, 6, 102]
print(heap)
```

```python
build_heap(heap)
print(heap)
```

```python
def heap_sort(lst):
    """ Perform heap sort by building a heap from the list and repeatedly retrieving the root element """
    n = len(lst)

    # Build Max Heap
    build_heap(lst)
    
    for i in reversed( range(n) ):
        #swap
        lst[i], lst[0] = lst[0], lst[i]

        # heapify root element again but only until the ith element
        heapify(lst, i, 0)
```

```python
heap = [100, 5, 3, 2, 8, 15, 6, 102]
print(heap)
heap_sort(heap)
print(heap)
```
