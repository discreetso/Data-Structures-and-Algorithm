```python
class Node:
    def __init__(self, data=None):
        self.val = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
```

```python
#The Push Operation
#    Push operation has two cases:
#        1. When there are no nodes
#        2. when there is already one or more nodes
```

```python
def push(self, val):
    new_node = Node(val)

    #if there is no node
    if self.head is None:
        self.head = new_node
        return

    #if there is one or more nodes
    last = self.head
    while last.next is not None:
        last = last.next
    last.next = new_node

LinkedList.push = push
```

```python
#Conversion to String
#    Python has a special function __str__. This is called whenever a cast to string is made. 
#    (These are called dunder (double underscore) functions)
```

```python
def __str__(self):
    ret_str = '['

    temp = self.head
    while temp is not None:
        ret_str += str(temp.val) + ', '
        temp = temp.next
    ret_str = ret_str.rstrip(', ')
    ret_str += ']'
    return ret_str

LinkedList.__str__ = __str__
```

```python
# Pop has two cases:
#     1. When there is only one node
#     2. When there are two or more nodes--in this case, we keep two pointers:
#        prev and temp. Move both until temp is the last. Then set next of prev to None.
```

```python
def pop(self):
    #if there is not any node
    if self.head is None:
        raise Exception('list is empty, No value, no popup')

    #if there is only one node
    if self.head.next is None:
        val = self.head.val
        self.head = None
        return val

    #if there is more than one node
    temp = self.head
    while temp.next is not None:
        prev = temp
        temp = temp.next
    val = temp.val
    prev.next = None
    return val

LinkedList.pop = pop
```

```python
# Insertion, again two cases
#    1. Insertion at index 0: new head, old head becomes next of this new head
#    2. Insertion at any other index: in this case, move prev and temp forward index times.
#       then insert new betweenm temp and prev
```

```python
def insert(self, index, val):
    new_node = Node(val)

    #insertion at index 0 is different
    if(index == 0):
        new_node.next = self.head
        self.head = new_node
        return

    #for other indices
    temp = self.head
    counter = 0
    while temp is not None and counter < index:
        prev = temp 
        temp = temp.next
        counter += 1
    prev.next = new_node
    new_node.next = temp

LinkedList.insert = insert
```

```python
# Remove operation: This is also the same:
#     1. If first node is present and same as val, remove it
#     2. otherwise, move prev and temp until temp points to the value. Set next of prev 
#        to next of temp.(temp is lost)
```

```python
def remove(self, val):
    temp = self.head

    #check first node
    if temp is not None:
        if temp.val == val:
            self.head = temp.next
            temp = None            #not necessary
            return

    while temp is not None:
        if temp.val == val:
            break
        prev = temp
        temp = temp.next
    if temp is None:              #not found
        return
    prev.next = temp.next         #just lose the reference to delete node

LinkedList.remove = remove
```

```python
# len: also have three cases
#    1. If there is no node, raise Exception: list is empty, length not calculated
#    2. if there is only one node: inc the counter once and return the counter
#    3. if there is more than one node: inc the counter until temp becomes None
#       and return the counter
```

```python
def len(self):
    #if there is no node
    if self.head is None:
        raise Exception('list is empty, length not calculated')
        
    temp = self.head
    counter = 0
    
    #if there is only one node
    if temp.next is None:
        counter += 1
        return counter

    #if there is more than one node    
    while temp is not None:
        counter += 1
        temp = temp.next
    return counter

LinkedList.len = len
```

```python
# get(index): also have three cases:
#     1. If there is no node; raise Exception List is empty
#     2. To get the value from index 0, just return value at temp
#     3. to get value from other than index 0; move forward temp and inc counter until temp becomes none or if counter equals index
```

```python
def get(self, index):
    #c1: if there is not any node
    if self.head is None:
        raise Exception('list is empty')

    temp = self.head
    #c2: if the index is first
    if index == 0:
        return temp.val

    #c3: index other than 0
    counter = 0               #temp already created and initialize above ;temp = self.head
    while temp is not None and counter<index:
        temp = temp.next
        counter += 1
    if temp is None:          #not found
        return
    return temp.val
LinkedList.get = get
```
