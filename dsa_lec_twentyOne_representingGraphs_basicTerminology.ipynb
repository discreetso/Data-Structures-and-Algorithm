```python
class Digraph:
    def __init__(self):
        self.g = {}

    def add_node(self, node):
        self.g[node] = []
```

```python
def add_edges(self, src, dest):
        # check if both (src, dest) exists
        if src not in self.g or dest not in self.g:
            raise ValueErorr('node does not exist')

        # we can also access the list and append node in one line of code (self.g[src].append(dest)) but we have to check whether
        # dest node already exists in the list or connected
    
        l = self.g[src]
        if dest in l:
            return
        l.append(dest)

Digraph.add_edges = add_edges
```
