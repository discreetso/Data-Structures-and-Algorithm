```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return '[' + str(self.x) + ', ' + str(self.y) + ']'
```

```python
p1 = Point()
print('p1.x = ', p1.x)

p2 = Point(2,4)
print('p2.x = ', p2.x)
```

```python
print(p2)
```

```python
#composition
class Shape:
    def __init__(self, points):
        self.points = points
```

```python
p3 = Point(4, 8)
p4 = Point(12, 16)
p5 = Point(20, 24)

p = [p3, p4, p5]
sh = Shape(p)
```

```python
#we can add methods even after class has been defined
def __str__(self):
    ret = ''

    for i in self.points:
        ret += str(i) + '-'

    return ret

Shape.__str__ = __str__
```

```python
print(sh)
```

```python
def print_points(self):
    for i in self.points:
        print(i)

Shape.print_points = print_points
```

```python
sh.print_points()
```

```python
#inheritance: inheritance syntax is also slightly different but quite easy

class Triangle(Shape):    #Triangle inherits from Shape
    pass                  #pass means i am not going to have anything in this block
```

```python
t = Triangle(p)
```

```python
t.print_points()
```

```python
def get_area(self):
    vertices = self.points
    n = len(vertices) # of corners
    a = 0.0
    for i in range(n):
        j = (i + 1) % n
        a += abs(vertices[i].x * vertices[j].y - vertices[i].y * vertices[j].x)
    return a / 2.0

Triangle.get_area = get_area
```

```python
t.get_area()
```

```python
#Access parent class overridden methods 
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.lenght + 2 * self.width

    def __str__(self):
        return 'L: ' + str(self.length) + ' W: ' + str(self.width)
```

```python
rect = Rectangle(2, 4)
```

```python
print(rect)
```

```python
#here we declare that the Square class inherits from the Rectangle class 
class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)
    def __str__(self):
        return 'Square: ' + super().__str__()
```

```python
square = Square(4)
```

```python
square.area()
```

```python
print(square)
```

```python
#Order of Growth
# Recall fibonacci series: 1,2,3,5,8,13 ....
#fib(n) = fib(n-1) + fib(n-2)
#we have two alternatives for computing fib(n)
```

```python
def fib1(n):
    if(n <= 1):
        return 1
    else:
        return fib1(n-1) + fib1(n-2)
```

```python
def fib2(n):
    a = b = 1

    for i in range(n):
        a, b = b, a + b
    return a 
```

```python
print('fib1: ')
for i in range(15):
    print(fib1(i), end = ',')

print('\n\n')

print('fib2: ')
for i in range(15):
    print(fib2(i), end = ',')
```

```python
import time
def compute_times(fn, limit):
    l = []
    for i in range(limit):
        start_time = int(round(time.time() * 1000))
        fn(i)
        end_time = int(round(time.time() * 1000))
        time_taken = end_time - start_time
        l.append(time_taken)
    return l
```

```python
limit = 30
fib1_times = compute_times(fib1, limit)
fib2_times = compute_times(fib2, limit)
```
