```python
class Reverse:
    def __init__(self, data):
        self.data = data
        self.index = len(data)

    def __iter__(self):
        return self

    def __next__(self):
        if self.index == 0:
            raise StopIteration
        self.index = self.index-1
        return self.data[self.index]
```

```python
rev = Reverse("spam")
iter(rev)
```

```python
for chars in rev:
    print(chars)
```

```python
def reverse(data):
    for index in range(len(data)-1, -1, -1):
        yield data[index]
```

```python
for char in reverse("golf"):
    print(char)
```
