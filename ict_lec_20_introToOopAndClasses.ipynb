```python
class Polygon:
    def __init__(self, no_of_sides):             # called a constructor
        print("Creating a polygon of sides: ", no_of_sides)
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def set_sides(self, sides):
        self.sides = sides

    def show_sides(self):
        for i in self.sides:
            print("side: ", i)
```

```python
# example of self keyword use
class Person:
    def __init__(self, name):
        name = name              # This does nothing 
        self.name = name         # this assigns to the instance
```

```python
# Another example: class demonstrating the use or benefit of self keyword
class Dog:
    def __init__(self, name):
        self.name = name         # stored in the object
        age = 5                  # local variable, discarded after init ends
    def sound(self):
        print("My name is ", self.name)
        sound = "Woof!"         # local variable, gone after speak ends
        print(sound)
```

```python
d = Dog("tommy")
print(d.name)
```

```python
# Now return to our Polygon class
q = Polygon(3)
p = Polygon(4)
print(q.n)
print(p.n)
print("Showing the number of sides of polygon p")
p.show_sides()
print("Setting and Showing the number of sides of polygon q")
q.set_sides([3, 4, 5])
print( q.show_sides() )
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self, 3)

    def get_area(self):
        a, b, c = self.sides
        # calculate the semi-perimeter
        s = (a + b + c) / 2
        area = (s * (s-a) * (s-b) * (s-c)**5)
        # just the formula
        return area
```

```python
t = Triangle()         # creating a polygon of sides 3
print(t.n)
t.set_sides([3, 4, 5])
t.show_sides()
print(t.get_area())
```

```python
class Rectangle(Polygon):
    def __init__(self):
        Polygon.__init__(self, 4)

    def set_sides(self, sides):
        assert sides[0] == sides[2]      # opposite sides must be equal
        assert sides[1] == sides[3]      # opposite sides must be equal
        self.sides = sides

    def get_area(self):
        area = self.sides[0] * self.sides[1]
        return area
```

```python
r = Rectangle()
print(r.n)
print("before setting sides: ")
r.show_sides()
r.set_sides([1,2,1,3])            # makes the opposite sides equal to get rid of the error
print("After setting sides: ")
r.show_sides()
print( r.get_area() )
```
